```python
data=df.copy()
```

```python
data.dtypes
```

```python
data.drop(['Title','Description','Rank', 'Actors'],axis=1, inplace =True)
```

```python
data.isnull().sum()
```

```python
data=data.dropna()
```

```python
data.isnull().sum()
```

```python
from sklearn.preprocessing import StandardScaler 
```

### Standardizing the numerical columns

```python
li=['Year', 'Runtime (Minutes)', 'Rating', 'Votes', 'Revenue (Millions)',
       'Metascore']
for i in li:
    scale = StandardScaler().fit(data[[i]])
    data[i] = scale.transform(data[[i]])
    data[i] = scale.transform(data[[i]])
```

```python
##encoding gud directors,i.e, more than 2 successful movies

pd.set_option('display.max_columns', None)
data=pd.get_dummies(data, 'good_directors')
data=pd.get_dummies(data, 'Genre')

```

```python
data
```

```python
from sklearn.model_selection import train_test_split 
from sklearn.metrics import classification_report,confusion_matrix 
from sklearn import metrics
from sklearn.svm import SVC
from sklearn.neighbors import KNeighborsClassifier
```

```python
X=data.drop(['Success'],axis=1)
X.shape
```

```python
y=data['Success']
y.shape
```

```python
y.value_counts().plot.bar()
```

### Balancing the result class using SMOTE

```python
from imblearn.over_sampling import SMOTE
sm = SMOTE(random_state = 2)
X, y = sm.fit_resample(X, y.ravel())
```

```python
sum(y == 0.0)   #BALANCED
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.3, random_state=42)
```

```python
clf = SVC(kernel='linear')
```

```python
clf.fit(X_train,y_train)
```

```python
y_pred= clf.predict(X_test)
```

```python
from sklearn import metrics
metrics.accuracy_score(y_test,y_pred)
```

```python
print(classification_report(y_test,y_pred))
```

### Thus with 84.5% accuracy our model is ready to be deployed
